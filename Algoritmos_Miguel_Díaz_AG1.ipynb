# Torres de Hanoi - Divide y venceras


```python
'''
Funcion de torre de Hanoi, se agrega un contador para mostrar el numero de movimientos
'''
def Torres_Hanoi(N, desde, hasta, contador_movimientos):
  #N - Nº de fichas
  #desde - torre inicial
  #hasta - torre fina
  if N==1 :
    print("Lleva la ficha desde " + str(desde) + " hasta " + str(hasta))
    contador_movimientos += 1
    return contador_movimientos

  else:
    contador_movimientos=Torres_Hanoi(N-1, desde, 6-desde-hasta,contador_movimientos)
    print("Lleva la ficha desde " + str(desde) + " hasta " + str(hasta))
    contador_movimientos += 1
    contador_movimientos=Torres_Hanoi(N-1, 6-desde-hasta,  hasta,contador_movimientos)
    return contador_movimientos

contador_movimientos =Torres_Hanoi(5, 1, 3,0)
print(f"Número total de movimientos: {contador_movimientos}")
```

# N Reinas - Vuelta Atrás()

```python
#N Reinas - Vuelta Atrás()
################################################################################

#Verifica que en la solución parcial no hay amenzas entre reinas
################################################################
def es_prometedora(SOLUCION,etapa):
################################################################
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

#Traduce la solución al tablero
################################################################
def escribe_solucion(S):
################################################################
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")



#Proceso principal de N-Reinas
################################################################
def reinas(N, solucion=[],etapa=0):
################################################################
### ....
  if len(solucion) == 0:         # [0,0,0...]
    solucion = [0 for i in range(N) ]

  for i in range(1, N+1):
    solucion[etapa] = i
    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

  solucion[etapa] = 0

reinas(8,solucion=[],etapa=0)
```

# Viaje por el rio - Programación dinámica

```python
#Viaje por el rio - Programación dinámica
################################################################################

TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]

#999 se puede sustituir por float("inf")


#Calculo de la matriz de PRECIOS y RUTAS
################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################

PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```

## Ejercicio extra:


Encontrar los dos puntos más cercanos

• Dado un conjunto de puntos se trata de encontrar los dos puntos más cercanos

• Guía para aprendizaje:

- Suponer en 1D, o sea, una lista de números: [3403, 4537, 9089, 9746, 7259, ….
- Primer intento: Fuerza bruta
- Calcular la complejidad. ¿Se puede mejorar?
- Segundo intento. Aplicar Divide y Vencerás
- Calcular la complejidad. ¿Se puede mejorar?
- Extender el algoritmo a 2D: [(1122, 6175), (135, 4076), (7296, 2741)…
- Extender el algoritmo a 3D.


```python
# Primero metodo para generar los puntos en 1D y 2D

import random as rd
LISTA_1D = [rd.randrange(1,10000) for x in range(1000)]
LISTA_2D = [(rd.randrange(1,10000), rd.randrange(1,10000)) for x in range(1000)]

print(LISTA_1D)
print(LISTA_2D)
```

```python
def puntos_cercanos_fuerza_bruta(points):
    distancia_minima = float('inf')
    par_cercano = None
    n = len(points)

    for i in range(n):
        for j in range(i + 1, n):
            distancia = abs(points[i] - points[j])
            if distancia < distancia_minima:
                distancia_minima = distancia
                par_cercano = (points[i], points[j])

    return par_cercano, distancia_minima


print(puntos_cercanos_fuerza_bruta(LISTA_1D))
```

La complejidad temporal es
\[O(n\^{2})\]

Mejorar usando Divide y Venceras:
## Divide y venceras:


```python


'''
El algoritmo en Divide y venceras se divide en:
-Ordenar los puntos
-Dividir el conjunto en mitades
-Encontrar los puntos mas cercanos en cada mitad de forma recursiva
-Encontrar los puntos mas cercanos que cruzan la linea divisoria
-Combinar los resultados
'''
def puntos_cercanos_divide_venceras(puntos):
    def puntos_cercanos_recursive(puntos_ordenados):
        n = len(puntos_ordenados)
        if n <= 3:
            return puntos_cercanos_fuerza_bruta(puntos_ordenados)

        mid = n // 2
        mitad_izq = puntos_ordenados[:mid]
        mitad_der = puntos_ordenados[mid:]

        (left_cercana, left_min_dist) = puntos_cercanos_recursive(mitad_izq)
        (right_cercana, right_min_dist) = puntos_cercanos_recursive(mitad_der)

        if left_min_dist < right_min_dist:
            min_distancia = left_min_dist
            cercana_pair = left_cercana
        else:
            min_distancia = right_min_dist
            cercana_pair = right_cercana

        cruzado_cercana, cruzado_min_dist = puntos_cercanos_par(puntos_ordenados, mid, min_distancia)

        if cruzado_min_dist < min_distancia:
            return cruzado_cercana, cruzado_min_dist
        else:
            return cercana_pair, min_distancia

    def puntos_cercanos_par(puntos_ordenados, mid, min_distancia):
        punto_medio = puntos_ordenados[mid]
        cruzado_puntos = [p for p in puntos_ordenados if abs(p - punto_medio) < min_distancia]

        cruzado_min_dist = min_distancia
        cruzado_cercana_pair = None
        len_cruzado_puntos = len(cruzado_puntos)

        for i in range(len_cruzado_puntos):
            for j in range(i + 1, min(i + 7, len_cruzado_puntos)):
                distance = abs(cruzado_puntos[i] - cruzado_puntos[j])
                if distance < cruzado_min_dist:
                    cruzado_min_dist = distance
                    cruzado_cercana_pair = (cruzado_puntos[i], cruzado_puntos[j])

        return cruzado_cercana_pair, cruzado_min_dist

    puntos_ordenados = sorted(puntos)
    return puntos_cercanos_recursive(puntos_ordenados)

print(puntos_cercanos_divide_venceras(LISTA_1D))
```

Extender a 2D el divide y venceras:
## Divide y venceras 2D

```python
import math

def distancia_2_puntos(p1, p2):
    return math.sqrt((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2)

def puntos_cercanos_2d(points):
    def puntos_cercanos_recursive(puntos_ordenados_x, puntos_ordenados_y):
        n = len(puntos_ordenados_x)
        if n <= 3:
            return puntos_cercanos_brute_force_2d(puntos_ordenados_x)

        mid = n // 2
        mitad_izq_x = puntos_ordenados_x[:mid]
        mitad_der_x = puntos_ordenados_x[mid:]
        punto_medio = puntos_ordenados_x[mid][0]

        mitad_izq_y = [p for p in puntos_ordenados_y if p[0] <= punto_medio]
        mitad_der_y = [p for p in puntos_ordenados_y if p[0] > punto_medio]

        (izq_cerca, izq_min_dist) = puntos_cercanos_recursive(mitad_izq_x, mitad_izq_y)
        (der_cerca, der_min_dist) = puntos_cercanos_recursive(mitad_der_x, mitad_der_y)

        if izq_min_dist < der_min_dist:
            min_distance = izq_min_dist
            puntos_cercanos = izq_cerca
        else:
            min_distance = der_min_dist
            puntos_cercanos = der_cerca

        cruzado_cerca, cruzado_min_dist = cerca_cruzado_par_2d(puntos_ordenados_y, punto_medio, min_distance)

        if cruzado_min_dist < min_distance:
            return cruzado_cerca, cruzado_min_dist
        else:
            return puntos_cercanos, min_distance

    def cerca_cruzado_par_2d(puntos_ordenados_y, punto_medio, min_distance):
        strip_points = [p for p in puntos_ordenados_y if abs(p[0] - punto_medio) < min_distance]

        cruzado_min_dist = min_distance
        cruzado_puntos_cercanos = None
        len_strip_points = len(strip_points)

        for i in range(len_strip_points):
            for j in range(i + 1, min(i + 7, len_strip_points)):
                distance = distancia_2_puntos(strip_points[i], strip_points[j])
                if distance < cruzado_min_dist:
                    cruzado_min_dist = distance
                    cruzado_puntos_cercanos = (strip_points[i], strip_points[j])

        return cruzado_puntos_cercanos, cruzado_min_dist

    def puntos_cercanos_brute_force_2d(points):
        min_distance = float('inf')
        puntos_cercanos = None
        n = len(points)

        for i in range(n):
            for j in range(i + 1, n):
                distance = distancia_2_puntos(points[i], points[j])
                if distance < min_distance:
                    min_distance = distance
                    puntos_cercanos = (points[i], points[j])

        return puntos_cercanos, min_distance

    puntos_ordenados_x = sorted(points, key=lambda p: p[0])
    puntos_ordenados_y = sorted(points, key=lambda p: p[1])
    return puntos_cercanos_recursive(puntos_ordenados_x, puntos_ordenados_y)

# Ejemplo de uso:
print(puntos_cercanos_2d(LISTA_2D))
```

Puntos en 3D


```python
def dist_3d(p1, p2):
    return math.sqrt((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2 + (p1[2] - p2[2]) ** 2)

def puntos_cercanos_3d(points):
    def puntos_cercanos_recursive(puntos_ordenados_x, puntos_ordenados_y, puntos_ordenados_z):
        n = len(puntos_ordenados_x)
        if n <= 3:
            return puntos_cercanos_brute_force_3d(puntos_ordenados_x)

        mid = n // 2
        mitad_izq_x = puntos_ordenados_x[:mid]
        mitad_der_x = puntos_ordenados_x[mid:]
        midpoint = puntos_ordenados_x[mid][0]

        mitad_izq_y = [p for p in puntos_ordenados_y if p[0] <= midpoint]
        mitad_der_y = [p for p in puntos_ordenados_y if p[0] > midpoint]

        mitad_izq_z = [p for p in puntos_ordenados_z if p[0] <= midpoint]
        mitad_der_z = [p for p in puntos_ordenados_z if p[0] > midpoint]

        (left_closest, left_min_dist) = puntos_cercanos_recursive(mitad_izq_x, mitad_izq_y, mitad_izq_z)
        (right_closest, right_min_dist) = puntos_cercanos_recursive(mitad_der_x, mitad_der_y, mitad_der_z)

        if left_min_dist < right_min_dist:
            min_distance = left_min_dist
            puntos_cercanos = left_closest
        else:
            min_distance = right_min_dist
            puntos_cercanos = right_closest

        cross_closest, cross_min_dist = cerca_cruzado_par_3d(puntos_ordenados_y, midpoint, min_distance)

        if cross_min_dist < min_distance:
            return cross_closest, cross_min_dist
        else:
            return puntos_cercanos, min_distance

    def cerca_cruzado_par_3d(puntos_ordenados_y, midpoint, min_distance):
        strip_points = [p for p in puntos_ordenados_y if abs(p[0] - midpoint) < min_distance]

        cross_min_dist = min_distance
        cross_puntos_cercanos = None
        len_strip_points = len(strip_points)

        for i in range(len_strip_points):
            for j in range(i + 1, min(i + 7, len_strip_points)):
                distance = dist_3d(strip_points[i], strip_points[j])
                if distance < cross_min_dist:
                    cross_min_dist = distance
                    cross_puntos_cercanos = (strip_points[i], strip_points[j])

        return cross_puntos_cercanos, cross_min_dist

    def puntos_cercanos_brute_force_3d(points):
        min_distance = float('inf')
        puntos_cercanos = None
        n = len(points)

        for i in range(n):
            for j in range(i + 1, n):
                distance = dist_3d(points[i], points[j])
                if distance < min_distance:
                    min_distance = distance
                    puntos_cercanos = (points[i], points[j])

        return puntos_cercanos, min_distance

    puntos_ordenados_x = sorted(points, key=lambda p: p[0])
    puntos_ordenados_y = sorted(points, key=lambda p: p[1])
    puntos_ordenados_z = sorted(points, key=lambda p: p[2])
    return puntos_cercanos_recursive(puntos_ordenados_x, puntos_ordenados_y, puntos_ordenados_z)

# Ejemplo de uso:
points = [(1122, 6175, 345), (135, 4076, 456), (7296, 2741, 567)]
print(puntos_cercanos_3d(points))
```
